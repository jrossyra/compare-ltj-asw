Plot the Landscapes for each lag from the 50us trajectory

```python
plt.close()
for lag, res in tica_data_Ca.items():

    fig, axes = plt.subplots(2, 2, figsize=(12, 8))

    pyemma.plots.plot_free_energy(*res[:, :2].T, ax=axes[0][0], cmap="RdYlGn_r")
    pyemma.plots.plot_free_energy(*res[:, [0, 2]].T, ax=axes[1][0], cmap="RdYlGn_r")
    pyemma.plots.plot_free_energy(*res[:, [2, 1]].T, ax=axes[0][1], cmap="RdYlGn_r")

    axes[1][0].set_xlabel("TIC 1 Coordinate")
    axes[0][0].set_ylabel("TIC 2 Coordinate")
    axes[1][0].set_ylabel("TIC 3 Coordinate")
    axes[0][1].set_xlabel("TIC 3 Coordinate")

    fig.delaxes(axes[1][1])

    plt.suptitle(
        "Landscape w/ lag: %.2f" % float(lag / step_per_ns[label_longtraj])
        + " ns using features C"
        + r"$\alpha$ 1/d",
        fontsize=16,
    )
    plt.tight_layout()
    plt.subplots_adjust(top=0.92)
    plt.savefig("analyses/ltj/tica-Calpha-010212landscapes.png", dpi=600)
```
