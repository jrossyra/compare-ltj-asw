```python
total = len(good_lag_ones)*(len(epochsizes)+1)
howmany = 0

for feat,wnm,setup in good_lag_ones:
    
    dds = da / wnm
    dtica = dds / "tica"
    ftica = dtica / "results.pyemma"

    dataset = all_models[feat][wnm]

    setup["byepochsize"] = list()
    
    thisone = _model_name(feat, setup["kwargs"])
    fbes = ftica.with_suffix(".%s"%thisone)
    
    if wnm == "ltj":
        _epochsizes = epochsizes+[-1]
        
    else:
        _epochsizes = aswa_tools.workloadbounds(
            dataset)[1:]

    print(colorama.Back.YELLOW)
    print(wnm)
    print(colorama.Back.RED)
    print(setup)
    print(thisone)
    print(dataset)
    
    for es in _epochsizes:
        
        howmany += 1
        
        print(colorama.Back.CYAN)
        print("Doing the lower data TICA calcs")
        print(colorama.Back.RESET)
        print(colorama.Fore.GREEN)
        print("On to number %d/%d"%(
            howmany,total))
        
        if es > 0:
            
            _fbes = Path(
                str(fbes) + \
                "-byepochsize-%dns" % es)
            
            print("created the path")
            print(_fbes)

            if _fbes.is_file():
                
                print(
        "found existing byepochsize file")
                
                result = pyemma.load(_fbes)
                
            else:
                
                print(colorama.Fore.BLACK)
                print(colorama.Back.MAGENTA)
                print(
        "calculating TICA by epochsize")
                
                result = coor.tica(
                    traj_features[(feat,wnm)][:es],
                    setup["kwargs"]["lag"],
                    reversible=False)
                
                result.save(
                    _fbes,
                    save_streaming_chain=True)
                
            setup["byepochsize"].append(
                (es, result))

        else: # Use TICA from full data
            
            print("found it longerone")
            
            result = setup["result"]
            
            setup["byepochsize"].append(
                (es, result))
            
        try:
            print(colorama.Back.BLUE)
            print(colorama.Fore.BLACK)
            print(pformat([ti.shape
                for ti in result.get_output(
                list(range(n_tica_dim)))]))

        except:

            print(colorama.Back.YELLOW)
            print(colorama.Fore.BLACK)
            print((
        "Something wrong with this model\n",
        " - trying to print TICA"
        " trajectories shapes"))

            print(pformat([ti.shape
                for ti in result.get_output()]))
            

print(colorama.Back.LIGHTBLACK_EX + \
      colorama.Fore.BLUE + \
      "  DONE LOADING DATA  ")
```
