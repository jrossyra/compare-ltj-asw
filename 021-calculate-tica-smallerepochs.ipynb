```python
for feat, wname, setup in good_lag_ones:
    
    dds = da/wname
    dtica = dds/"tica"
    ftica = dtica/"results.pyemma"
    dataset = all_models[feat][wname]
    setup["byepochsize"] = list()
    
    if wname == "ltj":
        _epochsizes = epochsizes
    else:
        _epochsizes = aswa_tools.workloadbounds(
            dataset)[1:]
        
    for es in _epochsizes:
            
        besetup = dict()
        besetup["epochsize"] = es
        besetup["kwargs"] = setup["kwargs"]
        besetup["timer"] = timer = Timer(
            description=True, verbose=True)
        besetup["input"] = inp = dataset[
            "feat_trajs"][:es]
        besetup["hash"] = thisone = _thisone_tica(
            feat, besetup)
        
        fbes = ftica.with_suffix(f".{thisone}")

        if fbes.is_file():
            print(f"{is1}Loading File:{rreset} {thisone}")
            
            with timer(yellowback+blacktxt+time_loadfile+rreset):
                result = pyemma.load(fbes)
                
        else:
            print(f"{is1}Calculating & Saving:{rreset} {thisone}")
            
            try:
                with timer(yellowback+blacktxt+time_calculation+rreset):
                    result = coor.tica(inp, **besetup["kwargs"])

                result.save(fbes, save_streaming_chain=True)
            
            except:
                print(yellowback+blacktxt+"Calculation failed!"+rreset)
                print(er1+traceback.format_exc()+rreset)
                result = False
        
        besetup["result"] = result
        setup["byepochsize"].append(besetup)


print(dn1+"  DONE LOADING DATA  ")
```
